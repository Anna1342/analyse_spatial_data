2. We print the crs. The epsg stands for European Petroleum Survey Group. It is a standard coding system for geographic coordinate systems and projections used in mapping and spatial data applications. EPSG code defines parameters like the coordinate system, projection method, and datum, which are essential for ensuring that spatial data aligns correctly across various sources and applications.
The CRS matters because it defines how the two-dimensional, flat map relates to real locations on the Earth’s surface. Without a common CRS, data from different sources can’t be accurately aligned or analyzed together.

5. We build a square of relevant dimensions and use the intersect method.

```python
from matplotlib.patches import Rectangle
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='lightgray', edgecolor='black', alpha=0.5)
cities_gdf.plot(ax=ax, color='blue', markersize=5, label='City Points')
rect_x = 4.5  # Longitude of bottom-left corner
rect_y = 44.5  # Latitude of bottom-left corner
rect_width = 3.0  # Width of the rectangle (e.g., degrees or projection units)
rect_height = 3.0  # Height of the rectangle

rect = Rectangle((rect_x, rect_y), rect_width, rect_height, linewidth=2, edgecolor='red', facecolor='none')
ax.add_patch(rect)
ax.set_xlim(gdf.total_bounds[[0, 2]]) 
ax.set_ylim(gdf.total_bounds[[1, 3]]) 
plt.show()
```

Here we just modify the chunk of code to plot only the region enclosed by the red rectangle. Instead of using the  bounds of gdf for ax.set_xlim() and ax.set_ylim(), we set them to match the rectangle's coordinates. 

```python
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='lightgray', edgecolor='black', alpha=0.5)
cities_gdf.plot(ax=ax, color='blue', markersize=5, label='City Points')
rect_x = 4.5  
rect_y = 44.5 
rect_width = 3.0  
rect_height = 3.0  

rect = Rectangle((rect_x, rect_y), rect_width, rect_height, linewidth=2, edgecolor='red', facecolor='none')
ax.add_patch(rect)
ax.set_xlim(rect_x, rect_x + rect_width)
ax.set_ylim(rect_y, rect_y + rect_height)

plt.show()
```

6. We are calculating the centroid of two districts to further calculate the distance between them and show it on the plot. 

```python
gdf.crs
gdf = gdf.to_crs(epsg=32631)
```

```python
print(gdf[['ID', 'centroid']].head())  
```

```python
gdf['centroid'] = gdf.geometry.centroid
centroid1 = gdf.loc[1, 'centroid']  
centroid2 = gdf.loc[4, 'centroid']  
print(centroid1)
print(centroid2)
```

```python
distance = centroid1.distance(centroid2)
print(distance/1000)
```

We plot the centroids on the map

```python
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='lightgrey', edgecolor='black')
ax.scatter(centroid1.x, centroid1.y, color='black', s=100, zorder=2)
ax.scatter(centroid2.x, centroid2.y, color='black', s=100, zorder=2)
ax.plot([centroid1.x, centroid2.x], [centroid1.y, centroid2.y], color='red', linestyle='-', linewidth=2)
midpoint = (centroid1.x + centroid2.x) / 2, (centroid1.y + centroid2.y) / 2
ax.legend(title=f'Distance: {distance / 1000:.2f} km')
```

5.3 Digital Elevation Models
We are going to create an interactive map with raster 

```python
import rasterio
import numpy as np
import plotly.graph_objects as go
path = "C:/Users/annas/Documents/python/output_SRTMGL1.tif"
with rasterio.open(path) as dataset:
    dem = dataset.read(1)
    dem = np.where(dem == dataset.nodata, np.nan, dem)
str(dem)
```

```python
height, width = dem.shape
x = np.arange(0, width)
y = np.arange(0, height)
X, Y = np.meshgrid(x, y)
```

```python
fig = go.Figure(data=[go.Surface(z=dem, x=X, y=Y, colorscale='Viridis', lighting=dict(ambient=0.3, diffuse=0.5, specular=1), lightposition=dict(x=100, y=200, z=300))])
fig.update_layout(title='3D Visualization',
                  scene=dict(
                      xaxis=dict(title='Width', titlefont=dict(size=13), gridcolor='white', showbackground=True, zerolinecolor='white'), 
                      yaxis=dict(title='Height', titlefont=dict(size=13), gridcolor='white', showbackground=True, zerolinecolor='white'),
                      zaxis=dict(title='Elevation', titlefont=dict(size=13), gridcolor='white', showbackground=True, zerolinecolor='white'),  
                      aspectratio=dict(x=1, y=1, z=0.5),
                      aspectmode='manual'))
fig.show()
```
